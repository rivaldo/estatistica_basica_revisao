```python
import pandas as pd
dados_escola = pd.read_csv('./censo_estadual_2007_a_2022.csv', sep=',', encoding='latin-1', low_memory=False)
dados_escola.drop(columns='Unnamed: 0', inplace=True)
dados_escola['NU_ANO_CENSO'] = pd.to_datetime(dados_escola['NU_ANO_CENSO'], format='%Y')
dados_escola.head()
```

```python
dados_escola['NU_ANO_CENSO'].unique()
```

```python
frequencia = dados_escola['NU_ANO_CENSO'].value_counts()
```

```python
numero_escolas = pd.DataFrame({
    'Número de Escolas - Rede Estadual': frequencia
})
numero_escolas.rename_axis('Ano do Censo', inplace=True)
numero_escolas['Diminuição no Número de Escolas'] = abs(numero_escolas.diff())
numero_escolas
```

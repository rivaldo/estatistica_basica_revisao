```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import pandas as pd
dados_escola = pd.read_csv('./censo_estadual_2007_a_2022.csv', sep=',', encoding='latin-1', low_memory=False)
dados_escola.drop(columns='Unnamed: 0', inplace=True)
dados_escola['NU_ANO_CENSO'] = pd.to_datetime(dados_escola['NU_ANO_CENSO'], format='%Y')
dados_escola.head()
```

## Quantitativo de matrículas 

```python
numero_matriculas = dados_escola.groupby('NU_ANO_CENSO')[['QT_MAT_BAS', 'QT_MAT_FUND','QT_MAT_MED', 'QT_MAT_PROF', 'QT_MAT_PROF_TEC', 'QT_MAT_BAS_FEM', 'QT_MAT_BAS_MASC']].sum()
numero_matriculas
```

```python
numero_matriculas.rename_axis('Ano do Censo', inplace=True)
numero_matriculas.rename(columns={
    'QT_MAT_BAS':       'Quant. de Mat. Educ. Basica', 
    'QT_MAT_FUND':      'Quant. de Mat. Fundamental',	
    'QT_MAT_MED':       'Quant. de Mat. Ensino Médio',
    'QT_MAT_PROF':      'Quant. de Mat. Ensino Profissional'	,
    'QT_MAT_PROF_TEC':  'Quant. de Mat. Ensino Profissional Técnico',
    'QT_MAT_BAS_FEM':   'Quant. de Mat. Ensino Básico Feminino',
    'QT_MAT_BAS_MASC':  'Quant. de Mat. Ensino Básico Masculino'
    },
    inplace=True
)
numero_matriculas
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(numero_matriculas.index, numero_matriculas['Quant. de Mat. Educ. Basica'], lw=3, color='g', label='Educaçao Básica - Total de Matriculas')
ax.plot(numero_matriculas.index, numero_matriculas['Quant. de Mat. Fundamental'], lw=3, color='r', label='Educaçao Fundamental - Total de Matriculas')
ax.plot(numero_matriculas.index, numero_matriculas['Quant. de Mat. Ensino Médio'], lw=3, color='b', label='Ensino Médio - Total de Matriculas')
ax.set_title('Comportamento do número de matriculas \nperíodo de 2007 a 2022', fontsize=18, loc='left' )
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12,)
ax.xaxis.set_tick_params(labelsize=12)
#plt.grid(linestyle='--')
plt.legend()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(numero_matriculas.index, numero_matriculas['Quant. de Mat. Educ. Basica'], lw=3, color='g', label='Educação Básica - Total Geral')
ax.plot(numero_matriculas.index, numero_matriculas['Quant. de Mat. Ensino Básico Feminino'], lw=3, color='r', label='Educação Básica - Total Feminino')
ax.plot(numero_matriculas.index, numero_matriculas['Quant. de Mat. Ensino Básico Masculino'], lw=3, color='b', label='Educação Básica - Total Masculino')
ax.set_title('Comportamento das matriculas segundo o genero dos alunos\nperíodo de 2007 a 2022', fontsize=18, loc='left' )
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12,)
ax.xaxis.set_tick_params(labelsize=12)
#plt.grid(linestyle='--')
plt.legend()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
numero_matriculas_cor_raça = dados_escola.groupby('NU_ANO_CENSO')[['QT_MAT_BAS_ND', 'QT_MAT_BAS_BRANCA', 'QT_MAT_BAS_PRETA', 'QT_MAT_BAS_PARDA', 'QT_MAT_BAS_AMARELA', 'QT_MAT_BAS_INDIGENA']].sum()
numero_matriculas_cor_raça.rename_axis('Ano do Censo', inplace=True)
numero_matriculas_cor_raça.rename(columns={
    'QT_MAT_BAS_ND':       'Quant. de Mat. Raça Não Declarada', 
    'QT_MAT_BAS_BRANCA':      'Quant. de Mat. Raça Branca',	
    'QT_MAT_BAS_PRETA':       'Quant. de Mat. Raça Preta',
    'QT_MAT_BAS_PARDA':      'Quant. de Mat. Raça Parda'	,
    'QT_MAT_BAS_AMARELA':  'Quant. de Mat. Raça Amarela',
    'QT_MAT_BAS_INDIGENA':   'Quant. de Mat. Raça Indigena',
    },
    inplace=True
)
numero_matriculas_cor_raça
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(numero_matriculas_cor_raça.index, numero_matriculas_cor_raça['Quant. de Mat. Raça Não Declarada'], lw=3, color='grey', label='Educação Básica - Raça não declarada')
ax.plot(numero_matriculas_cor_raça.index, numero_matriculas_cor_raça['Quant. de Mat. Raça Branca'], lw=3, color='red', label='Educação Básica - Raça branca')
ax.plot(numero_matriculas_cor_raça.index, numero_matriculas_cor_raça['Quant. de Mat. Raça Preta'], lw=3, color='black', label='Educação Básica - Raça preta')
ax.plot(numero_matriculas_cor_raça.index, numero_matriculas_cor_raça['Quant. de Mat. Raça Parda'], lw=3, color='green', label='Educação Básica - Raça parda')
ax.plot(numero_matriculas_cor_raça.index, numero_matriculas_cor_raça['Quant. de Mat. Raça Amarela'], lw=3, color='yellow', label='Educação Básica - Raça amarela')
ax.plot(numero_matriculas_cor_raça.index, numero_matriculas_cor_raça['Quant. de Mat. Raça Indigena'], lw=3, color='blue', label='Educação Básica - Raça indigena')
ax.set_title('Comportamento das matriculas segundo Cor/Raça dos alunos\nperíodo de 2007 a 2022', fontsize=18, loc='left' )
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12,)
ax.xaxis.set_tick_params(labelsize=12)
#plt.grid(linestyle='--')
plt.legend()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

## Informações Relativas Segundo a Faixa Etária dos Alunos

```python
numero_matriculas_faixa_etaria = dados_escola.groupby('NU_ANO_CENSO')[['QT_MAT_BAS_0_3', 'QT_MAT_BAS_4_5', 'QT_MAT_BAS_6_10', 'QT_MAT_BAS_11_14', 'QT_MAT_BAS_15_17', 'QT_MAT_BAS_18_MAIS']].sum()
numero_matriculas_faixa_etaria.rename_axis('Ano do Censo', inplace=True)
numero_matriculas_faixa_etaria.rename(columns={
    'QT_MAT_BAS_0_3':           'Quant. de Mat. de 0 a 3 anos', 
    'QT_MAT_BAS_4_5':           'Quant. de Mat. de 4 a 5 anos',	
    'QT_MAT_BAS_6_10':          'Quant. de Mat. de 6 a 10 anos',
    'QT_MAT_BAS_11_14':         'Quant. de Mat. de 11 a 14 anos'	,
    'QT_MAT_BAS_15_17':         'Quant. de Mat. de 15 a 17 anos',
    'QT_MAT_BAS_18_MAIS':       'Quant. de Mat. 18 anos ou mais',
    },
    inplace=True
)
numero_matriculas_faixa_etaria
```
